```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_data.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df= df.drop("Unnamed: 0",axis = 1)
df.head()
```

# 1. Gender Distribution

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender",palette=["skyblue", "teal"])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Gender Distribution")
plt.show()
```

**from the above chart we have analysed that:
the number of females in the data is more than the number of males**

# 2. Parent's education impact on different scores

```python
groupby = df.groupby ("ParentEduc").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(groupby)
```

```python
plt.figure(figsize = (9,6))
sns.heatmap(groupby, annot = True)
plt.title("Parents Education on Students Score")
plt.show()
```

**From the above chart we have concuded that the education of the paernts have a huge impact on their scores**

# 3. Parent's Marital Status impact on different scores

```python
groupby1 = df.groupby ("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":"mean","WritingScore":"mean"})
print(groupby1)
```

```python
plt.figure(figsize = (9,6))
sns.heatmap(groupby1, annot = True)
plt.title("Parents Martial Status on Students Score")
plt.show()
```

**From the above chart we have concluded that there is no/neglegible impact on the student score due to their parent's martial status**

# 5.Distribution of Ethnic Groups

```python
Group_A = df.loc[(df['EthnicGroup'] == "group A")].count()
Group_B = df.loc[(df['EthnicGroup'] == "group B")].count()
Group_C = df.loc[(df['EthnicGroup'] == "group C")].count()
Group_D = df.loc[(df['EthnicGroup'] == "group D")].count()
Group_E = df.loc[(df['EthnicGroup'] == "group E")].count()

label = ["group A","group B","group C","group D","group E"]
myList = [Group_A["EthnicGroup"],Group_B["EthnicGroup"],Group_C["EthnicGroup"],Group_D["EthnicGroup"],Group_E["EthnicGroup"]]
print(myList)
plt.pie(myList, labels = label, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()                                                                                                  
```
